```python
base_olympic_read = pd.read_csv("output.csv")
```

```python
base_olympic_read
```

```python
year = base_olympic_read["Year"].unique()
country = base_olympic_read["Country"].unique()
```

```python
base_olympic_read["Year"].unique()
```

```python
for varcountry in country:
    tmpvalue = 0
    for varyear in year:
        for i in range (0, len(base_olympic_read)):
            if (base_olympic_read["Country"][i] == varcountry and base_olympic_read["Year"][i] == varyear):
                tmpmedal = base_olympic_read["Medal"][i]
                tmpmedal2 = tmpmedal + tmpvalue
                tmpvalue = tmpmedal2
                base_olympic_read['Total_Medal'][i] = tmpmedal2
                
base_olympic_read
base_olympic_read.to_csv(r'finale_base_barchart.csv', index=False)
```
