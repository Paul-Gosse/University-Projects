```python
with open ("finale_base_barchart.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)
    data = []
    for row in reader:
        data.append({"date": row[0]+"-01-01", "name": row[1], "value": int(row[3])})
        
        
with open ("data_barchart_version2.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
with open ("finale_base_barchart.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)
    data = []
    nom = []
    for row in reader:
        if(row[1] not in nom):
            nom.append(row[1])
            data.append(row[1])
        
        
with open ("data_names_barchart.json", "w") as f:
    json.dump(data, f, indent=4)
```
