```python
import pandas as pd
import csv
```

```python
base_winter = pd.read_csv("winter.csv")
```

```python
base_summer = pd.read_csv("summer.csv")
```

```python
base_winter.pop('Year')
base_winter.pop('City')
base_winter.pop('Athlete')
base_winter.pop('Country')
base_winter.pop('Gender')
base_winter.pop('Medal')

base_winter
```

```python
output_winter = base_winter.drop_duplicates(subset = ["Event"])
output_winter.insert(loc=0, column = 'Season', value = 'Winter')
output_winter
```

```python
base_summer.pop('Year')
base_summer.pop('City')
base_summer.pop('Athlete')
base_summer.pop('Country')
base_summer.pop('Gender')
base_summer.pop('Medal')

base_summer
```

```python
output_summer = base_summer.drop_duplicates(subset = ["Event"])
output_summer.insert(loc=0, column = 'Season', value = 'Summer')
output_summer
```

```python
output_winter.to_csv(r'base_winter_sport.csv', index = False)
output_summer.to_csv(r'base_summer_sport.csv', index = False)
```

```python
output_winter_read = pd.read_csv("base_winter_sport.csv")
output_summer_read = pd.read_csv("base_summer_sport.csv")
```

```python
frames = [output_winter_read, output_summer_read]
output = pd.concat(frames)
output.to_csv(r'final_base_sports.csv', index=False)
```
