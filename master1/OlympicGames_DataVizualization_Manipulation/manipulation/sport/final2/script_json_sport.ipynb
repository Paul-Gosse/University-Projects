```python
with open ("final_base_sports.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)
    data = {"name": "flare", "children": []}
    season = ""
    sport_winter = []
    sport_summer = []
    discipline_winter = []
    discipline_summer = []
    cptWinter = -1
    cptSummer = -1
    tmpListWinter = []
    tmpListSummer = []
    for row in reader:
        if(row[0]!=season):
            season = row[0]
            data["children"].append({"name": row[0], "children": []})
            
        if(row[0] == "Winter" and row[2] not in discipline_winter):
            array = [row[1],row[2]]
            tmpArray = [row[1]]
            for elem in tmpArray:
                if elem not in tmpListWinter:
                    tmpListWinter.append(elem)
                    data["children"][0]["children"].append({"name": elem, "children": []})
            
            for elem2 in tmpArray:
                if elem2 in tmpListWinter:
                    cptWinter = tmpListWinter.index(elem2)
                    data["children"][0]["children"][cptWinter]["children"].append({"name": array[1], "children": []})
            discipline_winter.append(array[1])
            
            
            

        if(row[0] == "Winter"):
            array = [row[1],row[2],row[3]]
            tmpArray = [row[1]]
            tmpArray2 = [row[2]]
                                                    
            for elem2 in tmpArray:
                if elem2 in tmpArray:
                    cptWinter = tmpListWinter.index(elem2)
                    for elem3 in tmpArray2:
                        if elem3 in tmpArray2:
                            tmp = data["children"][0]["children"][cptWinter]["children"]
                            title = list(map(lambda x : x['name'], tmp))
                            for elem in title:
                                if(elem == elem3):
                                    cptWinter2 = title.index(elem)
                                    data["children"][0]["children"][cptWinter]["children"][cptWinter2]["children"].append({"name": array[2], "value": 1})
            
            
            
        if(row[0] == "Summer" and row[2] not in discipline_summer):
            array = [row[1],row[2]]
            tmpArray = [row[1]]
            for elem in tmpArray:
                if elem not in tmpListSummer:
                    tmpListSummer.append(elem)
                    data["children"][1]["children"].append({"name": elem, "children": []})
            
            for elem2 in tmpArray:
                if elem2 in tmpListSummer:
                    cptSummer = tmpListSummer.index(elem2)
                    data["children"][1]["children"][cptSummer]["children"].append({"name": array[1], "children": []})
            discipline_summer.append(array[1])
            
            
            
            
        if(row[0] == "Summer"):
            array = [row[1],row[2],row[3]]
            tmpArray = [row[1]]
            tmpArray2 = [row[2]]
                                                    
            for elem2 in tmpArray:
                if elem2 in tmpArray:
                    cptSummer = tmpListSummer.index(elem2)
                    for elem3 in tmpArray2:
                        if elem3 in tmpArray2:
                            tmp = data["children"][1]["children"][cptSummer]["children"]
                            title = list(map(lambda x : x['name'], tmp))
                            for elem in title:
                                if(elem == elem3):
                                    cptSummer2 = title.index(elem)
                                    data["children"][1]["children"][cptSummer]["children"][cptSummer2]["children"].append({"name": array[2], "value": 1})
```

```python
with open ("data_json_sport.json", "w") as f:
    json.dump(data, f, indent=4)
```
