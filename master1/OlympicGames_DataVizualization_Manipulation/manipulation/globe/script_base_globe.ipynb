```python
with open ("host_city.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)
    data = []
    for row in reader:
        data.append({"city": row[0], "country": row[1], "season": row[2], "year": row[3], "latitude": row[4], "longitude": row[5]})
```

```python
with open ("data_host_cities.json", "w") as f:
    json.dump(data, f, indent=4)
```
